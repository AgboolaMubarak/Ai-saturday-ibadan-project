One of the most basic features we can extract is the number of words in each tweet. The basic intuition behind this is that generally, the negative sentiments contain a lesser amount of words than the positive ones.

```python
data['word_count']=data['Tweet'].apply(lambda x: len(str(x).split(" ")))
```

```python
data[['Tweet','word_count']].head()
```

 This feature is also based on the previous feature intuition. Here, we calculate the number of characters in each tweet. This is done by calculating the length of the tweet.

```python
data['char_count'] = data['Tweet'].str.len() ## this also includes spaces
data[['Tweet','char_count']].head()
```

We will also extract another feature which will calculate the average word length of each tweet.
Here, we simply take the sum of the length of all the words and divide it by the total length of the tweet:

```python
def avg_word(sentence):
    words = sentence.split()
    return (sum(len(word) for word in words)/len(words))

data['avg_word'] = data['Tweet'].apply(lambda x: avg_word(x))
data[['Tweet','avg_word']].head()
```

Generally, while solving an NLP problem, the first thing we do is to remove the stopwords. But sometimes calculating the number of stopwords can also give us some extra information which we might have been losing before.

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

data['stopwords'] = data['Tweet'].apply(lambda x: len([x for x in x.split() if x in stop]))
data[['Tweet','stopwords']].head()
```

# Basic Pre-processing

The first pre-processing step which we will do is transform our tweets into lower case. This avoids having multiple copies of the same words. For example, while calculating the word count, ‘Analytics’ and ‘analytics’ will be taken as different words.

```python
data['Tweet'] = data['Tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data['Tweet'].head()
```

The next step is to remove punctuation, as it doesn’t add any extra information while treating text data. 

```python
data['Tweet'] = data['Tweet'].str.replace('[^\w\s]','')
data['Tweet'].head()
```

stop words (or commonly occurring words) should be removed from the text data. For this purpose, we can either create a list of stopwords ourselves or we can use predefined libraries

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
data['Tweet'] = data['Tweet'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
data['Tweet'].head()
```

Previously, I just removed commonly occurring words in a general sense. I can also remove commonly occurring words from our text data First, let’s check the 10 most frequently occurring words in our text data then take call to remove or retain.

```python
freq= pd.Series(' '.join(data['Tweet']).split()).value_counts()[:10]
freq
```

Now, let’s remove these words as their presence will not of any use in classification of our text data.

```python
freq = list(freq.index)
data['Tweet'] = data['Tweet'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
data['Tweet'].head()
```

similarly, just as I removed the most common words, this time let’s remove rarely occurring words from the text. Because they’re so rare, the association between them and other words is dominated by noise.

```python
freqi = pd.Series(' '.join(data['Tweet']).split()).value_counts()[-10:]
freqi
```

```python
freqi= list(freqi.index)
data['Tweet'] = data['Tweet'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
data['Tweet'].head()
```

spelling correction is a useful pre-processing step because this also will help us in reducing multiple copies of words.

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
data['Tweet'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))#Stemming refers to the removal of suffices, like “ing”, “ly”, “s”, etc. by a simple rule-based approach. For this purpose, we will use PorterStemmer from the NLTK library.
```

Lemmatization is a more effective option than stemming because it converts the word into its root word, rather than just stripping the suffices. 

```python
from textblob import Word
data['Tweet'] = data['Tweet'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data['Tweet'].head()
```

# Advance Text Processing

Up to this point, I have done all the basic pre-processing steps in order to clean our data. Now, I can finally move on to extracting features using NLP techniques.

N-grams are the combination of multiple words used together. Ngrams with N=1 are called unigrams. Similarly, bigrams (N=2), trigrams (N=3) and so on can also be used.

```python
TextBlob(data['Tweet'][0]).ngrams(2)
```

Term frequency is simply the ratio of the count of a word present in a sentence, to the length of the sentence

```python
tf1 = (data['Tweet'][1:2]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

The intuition behind inverse document frequency (IDF) is that a word is not of much use to us if it’s appearing in all the documents.

```python
for i,word in enumerate(tf1['words']):
    tf1.loc[i, 'idf'] = np.log(data.shape[0]/(len(data[data['Tweet'].str.contains(word)])))

tf1
```

TF-IDF is the multiplication of the TF and IDF which we calculated above.

```python
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```

 I also decide to use  sklearn as separate function to directly obtain TF and IDF:

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1000, lowercase=True, analyzer='word',
 stop_words= 'english',ngram_range=(1,1))
train_vect = tfidf.fit_transform(data['Tweet'])
```
